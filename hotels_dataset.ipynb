```python
finalData = pd.DataFrame()
for j in range(0,6000,25):
  url = 'https://www.booking.com/searchresults.en-gb.html?ss=Bengaluru&offset={}&ssne=Bengaluru&ssne_untouched=Bengaluru&efdco=1&label=gen173nr-&sid=2f4d2a78c5e299d36f3bd289c267240e&aid=304142&lang=en-gb&sb=1&src_elem=sb&src=searchresults&dest_id=-2090174&dest_type=city&checkin=2025-06-18&checkout=2025-06-19&group_adults=2&no_rooms=1&group_children=0'.format(j)

  headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
  }

  Response=requests.get(url,headers=headers)
  soup = BeautifulSoup(Response.text,'lxml')
  hotel = soup.find_all('div',role="listitem")
  hotel_name = []
  hotel_location = []
  hotel_location_detail = []
  quality = []
  reviews = []
  rating = []
  price = []


  for i in hotel:
      hotel_name.append(i.find('div',class_="b87c397a13 a3e0b4ffd1").text.strip())
      hotel_location.append(i.find('span',class_="d823fbbeed f9b3563dd4").text.strip())
      hotel_location_detail.append(i.find('div',class_="fff1944c52").text.strip())
      quality_element = i.find('div',class_="f63b14ab7a f546354b44 becbee2f63")
      quality.append(quality_element.text.strip() if quality_element else '')
      reviews_element = i.find('div',class_="fff1944c52 fb14de7f14 eaa8455879")
      reviews.append(reviews_element.text.replace('reviews','') if reviews_element else '')
      rating_element = i.find('div',class_="f63b14ab7a dff2e52086")
      rating.append(rating_element.text.strip() if rating_element else '')
      price.append(i.find('a',href=True).get('href'))

  df = pd.DataFrame({'hotel_name':hotel_name,'hotel_location':hotel_location,'hotel_location_detail':hotel_location_detail,'quality':quality,'reviews':reviews,'rating':rating,'price':price})
  finalData = pd.concat([finalData,df],ignore_index=True)
```
